## Determine if the column is categorical or numerical

```python
unique_vals=[]

for c in df.columns:
    vals_cnt=df[c].value_counts().count()
    unique_vals.append(vals_cnt)

pd.DataFrame(unique_vals,index=df.columns,columns=['Total unique values'])
```

## **Machine Learning**

```python
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.metrics import accuracy_score

import warnings
warnings.filterwarnings('ignore')
```

```python
df.head()
```

```python
df2=pd.get_dummies(df,columns=categorical_cols[:-1] , drop_first=True,dtype=int)
```

```python
df2.head()
```

```python
x=df2.drop(['output'],axis=1)
y=df2['output']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
```

-----
## **Build Models**

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model=DecisionTreeClassifier()
dt_model.fit(X_train,y_train)
```

```python
y_pred=dt_model.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

---------
## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier()
rf_model.fit(X_train,y_train)
```

```python
y_pred=rf_model.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
# using grid search
rf2_model=RandomForestClassifier()
paramaters={
    'n_estimators':[10,15,20,30,40,50,100,150,200],
    'criterion':['gini','entropy'],
    'bootstrap':[True,False],
    'max_features':['Auto',None,0.2]
}

rf2_model_gs=GridSearchCV(rf2_model,param_grid=paramaters)
rf2_model_gs.fit(X_train,y_train)
```

```python
rf2_model_gs.best_params_
```

```python
f_rf_model=RandomForestClassifier(bootstrap=True,criterion='entropy',n_estimators=15,max_features=0.2)
f_rf_model.fit(X_train,y_train)
```

```python
y_pred=f_rf_model.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

-------
## Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
svm_model=SVC()
svm_model.fit(X_train,y_train)
```

```python
y_pred=svm_model.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

--------
## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model=LogisticRegression()
lr_model.fit(X_train,y_train)
```

```python
y_pred=lr_model.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

-------
## KNN

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
knn_model=KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train,y_train)
```

```python
y_pred=knn_model.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```
